## 2.0 Collect data

### 2.1 define time window

```python
from datetime import datetime, timedelta, timezone # Import timezone

end_date = datetime.now(timezone.utc) # Use timezone-aware UTC
start_date = end_date - timedelta(days=30)
print("Collecting from", start_date, "to", end_date)
```

### 2.2 Fetching top users

```python
import string
import time

# Your existing Mastodon setup
# mastodon = Mastodon(access_token=ACCESS_TOKEN, api_base_url='https://mastodon.social')

# Parameters

search_patterns = list(string.ascii_lowercase + string.digits)  # a–z, 0–9
delay = 2  # seconds between requests to avoid rate limiting

all_accounts = {}

print("Starting user collection from Mastodon using account_search...")

for pattern in search_patterns:
    print(f"Searching for users with pattern '{pattern}'...")
    results = mastodon.account_search(pattern)
    time.sleep(delay)

    for acc in results:
        user_id = acc["id"]
        featured_tags = mastodon.account_featured_tags(user_id)
        featured_names = [tag["name"] for tag in featured_tags] if featured_tags else []
        if user_id not in all_accounts:
            all_accounts[user_id] = {
                "user_id": user_id,
                "username": acc["username"],
                "display_name": acc.get("display_name", ""),
                "user_type": "bot" if acc.get("bot", False)
                             else ("verified" if acc.get("verified", False)
                             else "user"),
                "followers_count": acc.get("followers_count", 0),
                "following_count": acc.get("following_count", 0),
                "statuses_count": acc.get("statuses_count", 0),
                "created_at": acc.get("created_at", ""),
                "featured_tags": ", ".join(featured_names)
            }

print(f"✅ Collected {len(all_accounts)} unique users")

# Convert to DataFrame
df = pd.DataFrame(all_accounts.values())

# Save to CSV
output_file = "mastodon_users.csv"
df.to_csv(output_file, index=False)

print(f"Saved user data to {output_file}")
```
